Missing values for categorical data

```python
features_nan = [feature for feature in test.columns if test[feature].isnull().sum()>0 and test[feature].dtypes=='O']
for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(test[feature].isnull().mean(),4)))
```

Replacing missing values with new labels

```python
def replace_cat_features(test,features_nan):
    data = test.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data
test = replace_cat_features(test,features_nan)

test[features_nan].isnull().sum()
```

Missing values for numerical variables

```python
numerical_with_nan = [feature for feature in test.columns if test[feature].isnull().sum()>0 and test[feature].dtypes!='O']

for feature in numerical_with_nan:
    print("{}: {}% missing values".format(feature,np.round(test[feature].isnull().mean(),4)))
```

Replacing numerical missing values

```python
for feature in numerical_with_nan:
    median_value = test[feature].median()
    
#     test[feature +' nan'] = np.where(test[feature].isnull(),1,0)
    test[feature].fillna(median_value,inplace=True)
    
test[numerical_with_nan].isnull().sum()
```

## Numerical values

Since numrical values are skewed we will perform log normal distribution

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']
#######Features which are skewed and does not have any nan values
for feature in num_features:
    test[feature] = np.log(test[feature])
```

```python
test[num_features].head()
```

## Handling rare categorical feature

We will remove the categoricaal varaibles that are present less than 1% of the observations

```python
categorical_features = [feature for feature in test.columns if test[feature].dtypes=='O']
```

```python
categorical_features
```

```python
# for feature in categorical_features:
#     temp = test.groupby(feature)['SalePrice'].count()/len(test)
#     temp_df = temp[temp>0.01].index
#     test[feature] = np.where(test[feature].isin(temp_df),test[feature],'Rare_var')
```

```python
for feature in categorical_features:
    temp = test[feature].value_counts()/ test.shape[0]
    temp_df = temp[temp>0.01].index
    test[feature] = np.where(test[feature].isin(temp_df),test[feature],'Rare_var')
```
